```python
from google.colab import files

image_without_dirt = files.upload()
image_with_dirt = files.upload()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import os
import cv2

image_without_dirt_path: str = 'zoro.jpg'
image_with_dirt_path: str = 'Bola.jpg'

def show_image_with_adjusted_channels(readed_image_without_dirt) -> None:
  image_adjusted_channels = cv2.cvtColor(readed_image_without_dirt, cv2.COLOR_BGR2RGB)
  plt.imshow(image_adjusted_channels)
  plt.title('Image with adjusted channels.')
  plt.show()

def show_image_with_gray_scale(readed_image_without_dirt) -> None:
  image_gray_scale = cv2.cvtColor(readed_image_without_dirt, cv2.COLOR_BGR2GRAY)
  plt.imshow(image_gray_scale, cmap='gray')
  plt.title('Image with gray scale.')
  plt.show()

def show_image_in_each_rgb_scale(readed_image_without_dirt) -> None:
  image_rgb_scales = cv2.cvtColor(readed_image_without_dirt, cv2.COLOR_BGR2RGB)
  (r_channel, g_channel, b_channel) = cv2.split(image_rgb_scales)

  plt.imshow(r_channel, cmap='gray')
  plt.title('Image in the R channel.')
  plt.show()

  plt.imshow(g_channel, cmap='gray')
  plt.title('Image in the G channel.')
  plt.show()

  plt.imshow(b_channel, cmap='gray')
  plt.title('Image in the B channel.')
  plt.show()

def show_image_with_blur(readed_image_with_dirt):
  plt.imshow(readed_image_with_dirt, cmap='gray')
  plt.title('Original image with dirt.')
  plt.show()

  blur = cv2.blur(readed_image_with_dirt, (5, 5))
  plt.imshow(blur, cmap='gray')
  plt.title('Image with blur applied.')
  plt.show()

def show_image_with_median_blur(readed_image_with_dirt):
  plt.imshow(readed_image_with_dirt, cmap='gray')
  plt.title('Original image with dirt.')
  plt.show()

  median_blur = cv2.medianBlur(readed_image_with_dirt, 15)
  plt.imshow(median_blur, cmap='gray')
  plt.title('Image with median blur applied.')
  plt.show()

def show_image_with_thresholding(readed_image_without_dirt):
  image_rgb_scales = cv2.cvtColor(readed_image_without_dirt, cv2.COLOR_BGR2RGB)
  (_, _, b_channel) = cv2.split(image_rgb_scales)
  (T, bin) = cv2.threshold(b_channel, 100, 255, cv2.THRESH_BINARY)
  plt.imshow(bin, cmap='gray')
  plt.title('Image with threshold applied.')

if (os.path.exists(image_without_dirt_path) and os.path.exists(image_with_dirt_path)):
    readed_image_without_dirt = cv2.imread(image_without_dirt_path, cv2.IMREAD_COLOR)
    readed_image_with_dirt = cv2.imread(image_with_dirt_path)

    show_image_with_adjusted_channels(readed_image_without_dirt)
    print('\n')
    show_image_with_gray_scale(readed_image_without_dirt)
    print('\n')
    show_image_in_each_rgb_scale(readed_image_without_dirt)
    print('\n')
    show_image_with_blur(readed_image_with_dirt)
    print('\n')
    show_image_with_median_blur(readed_image_with_dirt)
    print('\n')
    show_image_with_thresholding(readed_image_without_dirt)
```
