```python
%matplotlib inline
from matplotlib import pyplot as plt
import os
import cv2

image_path: str = 'luffy.jpg'

if os.path.exists(image_path):
  luffy_image = cv2.imread(image_path, cv2.IMREAD_COLOR)
  luffy_image = cv2.cvtColor(luffy_image, cv2.COLOR_BGR2RGB)
  plt.imshow(luffy_image)
  plt.title('Luffy with original colors.')
  plt.show()
```
